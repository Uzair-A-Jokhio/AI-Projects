```python
## create a simple txt file

import os
os.makedirs("../data/text_files", exist_ok=True)
```

```python
sample_text = {
    "../data/text_files/python.txt":"""PYTHON: A BRIEF INTRODUCTION

Python is a high-level, interpreted, general-purpose programming language. It was first released in 1991 by Guido van Rossum.

KEY FEATURES:

1.  **Readability:** Python emphasizes code readability with its clean, whitespace-based syntax (indentation). This makes the code easier to learn, read, and maintain than many other languages.
2.  **Versatility (General Purpose):** It is used for a vast range of tasks, including web development (Django, Flask), data science (Pandas, NumPy), machine learning (PyTorch, TensorFlow), scripting, automation, and software testing.
3.  **Interpreted:** Python code is executed line-by-line by an interpreter, which allows for fast prototyping and debugging.
4.  **Large Standard Library:** Python comes with a massive collection of modules and functions (the standard library) that handle common programming tasks without needing to install external packages.

Python's simple structure and extensive ecosystem have made it one of the most popular programming languages globally for both beginners and experienced developers.""",

"../data/text_files/machine-learning.txt":"""MACHINE LEARNING (ML): A BRIEF INTRODUCTION

Machine Learning is a subfield of Artificial Intelligence (AI) that focuses on building systems that can learn from data, identify patterns, and make decisions with minimal human intervention.
CORE CONCEPT:
Instead of being explicitly programmed with rules to solve a problem (traditional programming), ML algorithms are fed large amounts of data and use statistical methods to discover relationships within that data.

MAIN TYPES OF ML:
1.  **Supervised Learning:** The model is trained on labeled data—meaning the input data is paired with the correct output (e.g., training a system with images labeled 'cat' or 'dog'). It is used for prediction and classification.
2.  **Unsupervised Learning:** The model is trained on unlabeled data and must find hidden patterns or structures on its own (e.g., grouping customers into segments). It is used for clustering and dimensionality reduction.
3.  **Reinforcement Learning (RL):** An agent learns by interacting with an environment, receiving rewards for correct actions and penalties for incorrect ones. It is often used in robotics and complex decision-making systems.

ML is the driving force behind modern technologies like recommendation systems, facial recognition, autonomous driving, and large language models (LLMs)."""
}

for filepath, content in sample_text.items():
    with open(filepath, "w", encoding="utf-8") as f:
        f.write(content)

print("Sample Text files created")
```

```python
### Textloader
from langchain_community.document_loaders import TextLoader

loader=TextLoader("../data/text_files/python.txt",encoding="utf-8")
document=loader.load()
print(document)
```

```python
## Directory Loader
from langchain_community.document_loaders import DirectoryLoader

dir_loader=DirectoryLoader(
    "../data/text_files",
    glob="**/*.txt",
    loader_cls= TextLoader,
    loader_kwargs={"encoding":"utf-8"},
    show_progress=False
)
txt_documents = dir_loader.load()
txt_documents
```

```python
from langchain_community.document_loaders import PyPDFLoader, PyMuPDFLoader

dir_loader = DirectoryLoader(
    "../data/pdf",
    glob="**/*.pdf",
    loader_cls=PyMuPDFLoader,
    show_progress=False,
)

pdf_documents = dir_loader.load()
pdf_documents
```
